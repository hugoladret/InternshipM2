### 2018-11-06 : Compte maintenant les fps

```python
import matplotlib.pyplot as plt
import imageio
from tqdm import tqdm
import skimage.transform
```

```python
filename = 'LifeOfBrian.mp4' #Do you find it risible when I say the name... Biggus Dickus ?
vid = imageio.get_reader(filename,  'ffmpeg')
```

```python
vid_array = [] #Two times faster as a list than as an array
for img in tqdm(range(vid.get_length())):
    try :
        img = vid.get_data(img)
        vid_array.append(img[:340,:,:]) #cutting off a black border
    except : 
        print('Empty frame error at frame %s' % img)
        pass
```

```python
plt.figure(figsize = (8,5))

plt.imshow(vid_array[200])
```

```python
down_vid_array = []
for img in tqdm(range(len(vid_array))):
    down_vid_array.append(skimage.transform.resize(vid_array[img][:,:,0],(9*4,16*4), mode = 'reflect'))
plt.imshow(down_vid_array[200], cmap = plt.cm.gray)
```

```python
import numpy as np
np.save('LifeOfBrian_%sFPS.npy' %vid.get_meta_data()['fps'], down_vid_array)
```
