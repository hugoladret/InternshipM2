# CAT Current Source Density stimulation

```python
import os
import numpy as np
import MotionClouds as mc
import datetime
import matplotlib.pyplot as plt
import csv
from tqdm import tqdm
import imageio
import warnings
from psychopy import visual, core, event
import random
```

```python
nbr_stim = 200 #repetition
stim_time = .100 #s
interstim_time = 1.5 #s

def get_keypress():
    keys = event.getKeys()
    if keys:
        return keys[0]
    else:
        return None
```

```python
# Initializing PsychoPy window
win = visual.Window(fullscr=True, color=[0.0075,0.0075,0.0075], winType='pygame', monitor='testMonitor', units='height')
mouse = event.Mouse(visible=False)

nparr = np.full((1920,1080), [0.0075])
bitmap1 = visual.ImageStim(win, size=3, units='height')
bitmap1.autolog = False
bitmap1.setImage(nparr)

bitmap1_1 = visual.ImageStim(win, size=3, units='height')
bitmap1_1.autolog = False
bitmap1_1.setImage(np.full((1920,1080), [0.0075]))


bitmap2 = visual.ImageStim(win, size=3, units='height')
bitmap2.autolog = False
bitmap2.setImage(np.full((1920,1080), [1]))

bitmap3 = visual.ImageStim(win, size=3, units='height')
bitmap3.autolog = False
bitmap3.setImage(np.full((1920,1080), [-1]))


for stimulations in range(nbr_stim):  
    clock = core.Clock()
    
    print('Trial no. ', stimulations)
    key = get_keypress()
    if key == 'escape' : 
        win.close()
        core.quit()
        
    #timing one presentation, it works perfectly fine
    while clock.getTime() < interstim_time+stim_time:
        if clock.getTime() < 0.05:  # fixation
            bitmap1.draw()
        elif .05 < clock.getTime() < interstim_time:  # fixation
            bitmap1_1.draw()
        else :
            if stimulations % 2 == 0 :
                bitmap2.draw()
            else :
                bitmap3.draw()
            #random.choice([bitmap2, bitmap3]).draw()
        win.flip()       

win.close()
```
