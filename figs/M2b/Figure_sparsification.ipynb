```python
from scipy.signal import savgol_filter
from scipy.stats import mannwhitneyu
for it, stk in enumerate([stack, stack2, stack3, stack4, stack5, stack6]) :
    plt.plot(stk[:,0])

    yhat = savgol_filter(stk[:,0], 9, 3)
    plt.plot(yhat)
    plt.show()
    
    middle = int(len(stk)/2)
    deep_fr = stk[:middle]
    superficial_fr = stk[middle:]
    
    fig = plt.figure(figsize=(10, 7.5))
    ax = plt.subplot(111)


    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)

    ax.set_xticklabels([])
    ax.set_xlim(-.5,5)
    
    plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) # labels along the bottom edge are off
    
    ax.set_ylabel('Somme P.A. par lectrode', fontsize=16)
    
    plt.bar([0,1],
       [np.sum(deep_fr), np.sum(superficial_fr)],
        width = 1,
       align = 'edge',
           color = '#3182bd')
    
    #plt.title( )
    print('MWU p-value %.4f' %mannwhitneyu(deep_fr[:,0], superficial_fr[:,0])[1])
    fig.savefig('Sparsification_stack%s' % it,
                format='svg', transparent=True, bbox_inches='tight')
    plt.show()
```
